```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
```

```python
#Question - 01
sample_rate = 1000
duration = 1.0
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

freq1 = 150
freq2 = 200

signal = np.sin(2 * np.pi * freq1 * t) + np.sin(2 * np.pi * freq2 * t)

plt.plot(t, signal, label='Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Signal')
plt.legend()
plt.show()
```

```python
#Question - 03
fft_result = np.fft.fft(signal)
freq = np.fft.fftfreq(len(fft_result), 1 / sample_rate)

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, signal, label='Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Signal')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(freq, np.abs(fft_result), label='Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Absolute Amplitude')
plt.title('Frequency Spectrum')
plt.xlim(0, sample_rate / 2)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Question - 04
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/lab7.csv')
df
```

```python
time = df['Time']
amplitude = df['Amplitude']
```

```python
from matplotlib import pyplot as plt
plt.plot(time,amplitude)
plt.title("time vs amplitude")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.legend(["amplitude"])
plt.show()
```

```python
#Question - 05
fft_result = np.fft.fft(amplitude)
freq = np.fft.fftfreq(len(fft_result), abs(time[1] - time[0]))

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(time, amplitude, label='Amplitude')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time vs. Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(freq, np.abs(fft_result), label='Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Absolute Amplitude')
plt.title('Frequency Spectrum')
plt.xlim(0, 1 / (2 * abs(time[1] - time[0])))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Question - 06
cutoff_frequency = 25
order = 4

b, a = butter(order, cutoff_frequency / (0.5 * len(time)), btype='high')

filtered_amplitude = lfilter(b, a, amplitude)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(time, amplitude, label='Original Amplitude')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time vs. Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(freq, np.abs(fft_result), label='Original Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Absolute Amplitude')
plt.title('Frequency Spectrum')
plt.xlim(0, 1 / (2 * abs(time[1] - time[0])))
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(time, filtered_amplitude, label='Filtered Amplitude', color='orange')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time vs. Filtered Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Question - 07
plt.figure(figsize=(12, 12))

plt.subplot(4, 1, 1)
plt.plot(time, amplitude, label='Original Amplitude')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time vs. Amplitude')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(freq, np.abs(fft_result), label='Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Absolute Amplitude')
plt.title('Frequency Spectrum')
plt.xlim(0, 1 / (2 * abs(time[1] - time[0])))
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(time, filtered_amplitude, label='Filtered Amplitude', color='orange')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Time vs. Filtered Amplitude')
plt.legend()

filtered_fft_result = np.fft.fft(filtered_amplitude)
plt.subplot(4, 1, 4)
plt.plot(freq, np.abs(filtered_fft_result), label='Filtered Frequency Spectrum', color='orange')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Absolute Amplitude')
plt.title('Filtered Frequency Spectrum')
plt.xlim(0, 1 / (2 * abs(time[1] - time[0])))
plt.legend()

plt.tight_layout()
plt.show()
```
