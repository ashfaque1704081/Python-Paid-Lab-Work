```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

projection = ccrs.Mercator()

fig, ax = plt.subplots(subplot_kw={'projection': projection})

ax.coastlines(resolution='10m', color='black')

glider_lat = [latitude[0], latitude[10], latitude[20]]
glider_lon = [longitude[0], longitude[10], longitude[20]]

ax.plot(glider_lon, glider_lat, 'ro', markersize=8, transform=ccrs.PlateCarree())

ax.set_extent([-180, 180, -75, 75], crs=ccrs.PlateCarree())

ax.gridlines(draw_labels=True, linestyle='--')

ax.set_title('Coastlines and Glider Positions (Mercator Projection)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```

```python
#Question-03
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a Mercator projection
projection = ccrs.Mercator()

# Create a figure and axis with the Mercator projection
fig, ax = plt.subplots(subplot_kw={'projection': projection})

# Add coastlines as a feature with a 10m resolution
ax.add_feature(cfeature.COASTLINE, linewidth=0.5, edgecolor='black')

# Define glider positions
glider_lat = [latitude[0], latitude[10], latitude[20]]
glider_lon = [longitude[0], longitude[10], longitude[20]]

# Plot the glider positions with red circles
ax.plot(glider_lon, glider_lat, 'ro', markersize=8, transform=ccrs.PlateCarree())

# Set the map extent and limits
ax.set_extent([-180, 180, -75, 75], crs=ccrs.PlateCarree())

# Add gridlines with labels and set the gridline color to yellow
ax.gridlines(draw_labels=True, linestyle='--')

# Set a title, xlabel, and ylabel
ax.set_title('Coastlines and Glider Positions (Mercator Projection)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Add a background image (Natural Earth)
ax.stock_img()

plt.show()
```

```python
#q7 Animation
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Load data from the netCDF file
data_file = "/content/drive/MyDrive/OS_OSNAP-WHOUC1_GLIDER_201506.nc"
dataset = nc.Dataset(data_file)
latitude = dataset.variables['LATITUDE'][:]
longitude = dataset.variables['LONGITUDE'][:]

# Create a scatter plot
fig, ax = plt.subplots()
sc = ax.scatter([], [])
line, = ax.plot([], [], 'r-', linewidth=0.5)  # Line for the glider path

# Set the plot limits
min_lat, max_lat = latitude.min(), latitude.max()
min_lon, max_lon = longitude.min(), longitude.max()
ax.set_xlim(min_lon, max_lon)
ax.set_ylim(min_lat, max_lat)

# Update function for the animation
def update(frame):
    glider_lat = latitude[:frame+1]  # Update the latitude path
    glider_lon = longitude[:frame+1]  # Update the longitude path
    sc.set_offsets(np.c_[glider_lon[-1], glider_lat[-1]])  # Set glider's current position
    line.set_data(glider_lon, glider_lat)  # Update the glider path line
    return sc, line

# Create the animation
num_frames = len(latitude)
ani = animation.FuncAnimation(fig, update, frames=num_frames, blit=True)

# Set up the animation writer
Writer = animation.writers['ffmpeg']
writer = Writer(fps=10, metadata=dict(artist='Me'), bitrate=1800)

# Save the animation as an mp4 file
animation_filename = 'glider_path_animation.mp4'
ani.save(animation_filename, writer=writer)

plt.show()
```
