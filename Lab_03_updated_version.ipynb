```python
#Extend the program to calculate sound speed from the data provided.
sound_speed=1448.96+4.591*temperature-0.05304*(temperature)**2+0.0002374*(temperature)**3+1.340*(salinity-35)+0.0163*(depth)\
            +0.000000000000747*(temperature*depth)**3
```

```python
def calculate_sound_speed(row):
    temperature = row['Temperature (C)']
    salinity = row['Salinity (PSU)']
    depth = row['Depth (m)']
    sound_speed = (
        1448.96 + 4.591 * temperature - 0.05304 * temperature**2 +
        0.0002374 * temperature**3 + 1.340 * (salinity - 35) +
        0.0163 * depth + 0.000000000000747 * (temperature * depth)**3
    )
    return sound_speed

df['Calculated Sound Speed (m/s)'] = df.apply(calculate_sound_speed, axis=1)
```
